```python
import pandas as pd
```

```python
data=pd.read_csv('/content/adult 3.csv')
data
```

```python
data['occupation'].replace({'?':'others'},inplace=True)
data['workclass'].replace({'?':'others'},inplace=True)
print(data.occupation.value_counts())
```

```python
print(data.occupation.value_counts())
print(data.workclass.value_counts())
```

```python
data=data[(data['workclass']!='Without-pay') & (data['workclass']!='Never-worked')]
print(data.workclass.value_counts())
```

```python
data.drop(columns=['relationship','marital-status','race'],inplace=True)
data
```

```python
from sklearn.preprocessing import OrdinalEncoder

education_order = [
    ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th',
     '10th', '11th', '12th', 'HS-grad', 'Assoc-voc', 'Assoc-acdm',
     'Some-college', 'Bachelors', 'Masters', 'Prof-school', 'Doctorate']
]
workclass_order = [
    ['others', 'Private', 'Self-emp-not-inc', 'Self-emp-inc',
     'Local-gov', 'State-gov', 'Federal-gov']
]

encoder1= OrdinalEncoder(categories=workclass_order)
data[['workclass']] = encoder1.fit_transform(data[['workclass']])
encoder2= OrdinalEncoder(categories=education_order)
data[['education']] = encoder2.fit_transform(data[['education']])
data
```

```python
from sklearn.preprocessing import LabelEncoder
enc1=LabelEncoder()
enc2=LabelEncoder()
enc3=LabelEncoder()
data['occupation']=enc1.fit_transform(data['occupation'])
data['gender']=enc2.fit_transform(data['gender'])
data['native-country']=enc3.fit_transform(data['native-country'])
data
```

```python
y=data['income']
x=data.drop(columns='income')
x
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

sc = MinMaxScaler()
l = sc.fit_transform(x)               # x should be your DataFrame of features
f = pd.DataFrame(l, columns=x.columns)
f

```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(f,y,test_size=0.2,random_state=38,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
kn=RandomForestClassifier()
kn.fit(xtrain,ytrain)
pred=kn.predict(xtest)
l=accuracy_score(ytest,pred)
l
```

```python
import pickle
with open('knf.pkl','wb') as file:
  pickle.dump(kn,file)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pickle

st.markdown(
    """
    <style>
    /* Main background */
    .stApp {
        background-color: orange;
    }
    </style>
    """,
    unsafe_allow_html=True
)
model_data = pickle.load(open("knf.pkl", "rb"))
kn = model_data['kn']
sc = model_data['sc']
enc1 = model_data['enc1']
enc2 = model_data['enc2']
enc3 = model_data['enc3']
encoder1 = model_data['encoder1']
encoder2 = model_data['encoder2']


st.markdown("<h1 style='color: black;'>Income Prediction App</h1>", unsafe_allow_html=True)


# Input from user
age = st.slider("Age", 0, 100)
workclass = st.selectbox("Workclass", ['others', 'Private', 'Self-emp-not-inc', 'Self-emp-inc', 'Local-gov', 'State-gov', 'Federal-gov'])
fnlwgt = st.number_input("FNLWGT", 0, 500000, 88901)
education = st.selectbox("Education", [
    'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th',
    '10th', '11th', '12th', 'HS-grad', 'Assoc-voc',
    'Assoc-acdm', 'Some-college', 'Bachelors', 'Masters',
    'Prof-school', 'Doctorate'
])
educational_num = st.number_input("Educational Number", 1, 16, 8)
occupation = st.selectbox("Occupation", ['Machine-op-inspct', 'Craft-repair', 'Exec-managerial', 'Sales', 'Prof-specialty'])  # Add more if needed
gender = st.selectbox("Gender", ['Male', 'Female'])
capital_gain = st.number_input("Capital Gain", 0, 100000, 0)
capital_loss = st.number_input("Capital Loss", 0, 100000, 2)
hours_per_week = st.number_input("Hours per Week", 1, 100, 50)
native_country = st.selectbox("Native Country", ['United-States', 'India', 'Mexico'])  # Add more if needed

if st.button("Predict"):
    # Prepare DataFrame
    arr = {
        'age': [age],
        'workclass': [workclass],
        'fnlwgt': [fnlwgt],
        'education': [education],
        'educational-num': [educational_num],
        'occupation': [occupation],
        'gender': [gender],
        'capital-gain': [capital_gain],
        'capital-loss': [capital_loss],
        'hours-per-week': [hours_per_week],
        'native-country': [native_country]
    }
    df = pd.DataFrame(arr)

    # Apply preprocessing
    df['gender'] = enc2.transform(df['gender'])
    df['native-country'] = enc3.transform(df['native-country'])
    df['occupation'] = enc1.transform(df['occupation'])
    df[['workclass']] = encoder1.transform(df[['workclass']])
    df[['education']] = encoder2.transform(df[['education']])
    df = sc.transform(df)

    # Predict
    prediction = kn.predict(df)[0]
    st.success(f"Prediction: {prediction}")
```
